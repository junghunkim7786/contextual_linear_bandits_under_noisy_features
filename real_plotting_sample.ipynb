```python
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
T = 10000

load = np.load('./real_outputs/result_ctr_20220819020826/arrays/ml100k_1660874938225_reward.npz')

clbef = load['clbbf']
oful  = load['oful']
random = load['random']

clbef_ctr  = np.cumsum(clbef)/(np.arange(T)+1)
oful_ctr   = np.cumsum(oful)/(np.arange(T)+1)
random_ctr = np.cumsum(random)/(np.arange(T)+1)

plt.plot(clbef_ctr,label='clbef')
plt.plot(oful_ctr,label='oful')
plt.plot(random_ctr,label='random')
plt.ylim(0,0.20)
plt.legend()
```
